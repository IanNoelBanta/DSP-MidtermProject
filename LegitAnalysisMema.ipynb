```python
import numpy as np
from scipy import stats
from scipy.fft import fft
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
excelFile = pd.read_excel("DSP-Data-1.xlsx")
copy = excelFile.copy()

excelFile
```

```python
data = pd.DataFrame({'var1': [1, 2, np.nan, 4, 5],
                     'var2': [2, np.nan, 4, 5, 6],
                     'var3': [np.nan, 3, 4, 5, np.nan],
                     'var4': [3, 4, 5, 6, 7],
                     'var5': [4, 5, 6, np.nan, 8],
                     'var6': [5, 6, 7, 8, 9]})
dataCopy = data.copy()
data
```

```python
missing_mask = dataCopy.isna()
```

```python
imputer = IterativeImputer(max_iter=4, random_state=0)

imputed_values = imputer.fit_transform(dataCopy)

dataCopy[missing_mask] = imputed_values[missing_mask]
```
