```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#wala lang
pd.set_option("display.max_rows", None)
```

```python
def getMean(x):
    return np.mean(x)

def getMedian(x):
    return np.median(x)

def getMode(x):
    vals, counts = np.unique(x, return_counts=True)
    mode = np.argwhere(counts == np.max(counts))
    return vals[mode].flatten()

def getVariance(x):
    return np.var(x)

def getStandardDeviation(x):
    # standard deviation represents noise and other interference.
    #  the standard deviation is not important in itself, but only in comparison to the mean.
    return np.std(x)

def getSignalToNoiseRatio(mean, std):
    return mean / std

def getCoefficientOfVariation(mean, std):
    #  a signal (or other group of measure values) with a CV of 2%, has an SNR of 50.  Better data means a higher value for the SNR and a lower value for the CV.
    return (std / mean) * 100
```

```python
# humidity
genHumiMean = getMean(humiGeneral)
genHumiMedian = getMedian(humiGeneral)
genHumiMode = getMode(humiGeneral)
genHumiVar = getVariance(humiGeneral)
genHumiSTD = getStandardDeviation(humiGeneral)
genHumiSTR = getSignalToNoiseRatio(genHumiMean, genHumiSTD)
genHumiCV = getCoefficientOfVariation(genHumiMean, genHumiSTD)
```

```python
def fftValue(data):
  value = fft(data) 
  freq = np.fft.fftfreq(len(fft(value))) 
  return freq, np.abs(value)

def fftPlotter(freq, value, title): 
  fig, ax = plt.subplots(figsize=(25, 5))
  
  ax.plot(freq, value, color='blue', label='FFT of Temperature', marker='o', markersize='2')
  
  ax.set_xlabel('Frequency (Hz)')
  ax.set_ylabel('Amplitude')
  
  # ax.set_ylim(0, np.max(value))
  # ax.axhline(color='k')
  
  ax.set_title(title + ' (Frequency Domain)')
  
  # Display the plot
  plt.grid(True)
  plt.legend()
  plt.show()
```

```python
freq, value = fftValue(tempDF) 
fftPlotter(freq, value, 'Temperature')
```

```python
freq, value = fftValue(humiDF) 
fftPlotter(freq, value, 'Humidity')
```
