```python
indexesOf5AMTo6AM = []
indexesOf1PMTo2PM = []
indexesOf10PMTo11PM = []

targets = ['05', '06', '13', '14', '22', '23']
    
for i in range(len(datetemphumiDF)):
    dateTimeSplit = datetemphumiDF['Date'][i].split('_')
    
    timeSplit = dateTimeSplit[1].split(':')
    hour = timeSplit[0]
    
    if hour in targets:
        if hour == '05'or hour == '06':
            indexesOf5AMTo6AM.append(i)
        if hour == '13' or hour == '14':
            indexesOf1PMTo2PM.append(i)
        if hour == '22' or hour == '23':
            indexesOf10PMTo11PM.append(i)
    
    
print(f'5AM to 6AM = {indexesOf5AMTo6AM}\n')
print(f'1PM to 2PM = {indexesOf1PMTo2PM}\n')
print(f'10PM to 11PM = {indexesOf10PMTo11PM}\n')
```

```python
indexesOf5AMTo6AMPerDay = []
tempOf5AMTo6AMPerDay = []
humiOf5AMTo6AMPerDay = []
datesOf5AMTo6AMPerDay = []

tempList = []
temperatures = []
humidity = []
datesList = []
prevDate = None

for indx in indexesOf5AMTo6AM:
    dateTimeSplit = datetemphumiDF['Date'][indx].split('_')
    dateSplit = dateTimeSplit[0].split('/')
    date = dateSplit[1]

    if date != prevDate and prevDate != None:
        indexesOf5AMTo6AMPerDay.append(tempList)
        tempOf5AMTo6AMPerDay.append(temperatures)
        humiOf5AMTo6AMPerDay.append(humidity)
        datesOf5AMTo6AMPerDay.append(datesList)
        
        tempList = []
        temperatures = []
        humidity = []
        datesList = []
    
    tempList.append(indx)
    temperatures.append(datetemphumiDF['Temperature (°C)'][indx])
    humidity.append(datetemphumiDF['Humidity (%)'][indx])
    datesList.append(date)
    
    prevDate = date
    
print(f'5AM to 6AM = {indexesOf5AMTo6AMPerDay}\n')
print(f'Dates of 5AM to 6AM = {datesOf5AMTo6AMPerDay}\n')
print(f'Temperatures of 5AM to 6AMM = {tempOf5AMTo6AMPerDay}\n')
print(f'Humidity of 5AM to 6AM = {humiOf5AMTo6AMPerDay}\n')
```
